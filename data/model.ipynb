```python
movies['Genre'] = movies['Genre'].str.replace('|', ' ')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')

vectors = cv.fit_transform(movies['Genre']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie_name):
    movie_index = movies[movies['MovieName'] == movie_name].index[0]
    
    distances = similarity[movie_index]
    
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(movies.iloc[i[0]].MovieName)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Clean genre
movies['Genre'] = movies['Genre'].str.replace('|', ' ')

tfidf = TfidfVectorizer(stop_words='english')

vectors = tfidf.fit_transform(movies['Genre']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie_name):
    movie_name = movie_name.lower()
    
    movies['MovieName'] = movies['MovieName'].str.lower()
    
    if movie_name not in movies['MovieName'].values:
        return ["Movie not found!"]
    
    movie_index = movies[movies['MovieName'] == movie_name].index[0]
    
    distances = similarity[movie_index]
    
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    result = []
    for i in movie_list:
        result.append(movies.iloc[i[0]].MovieName)
    
    return result
```

```python
recommend("Toy Story (1995)")
```
